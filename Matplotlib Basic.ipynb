### see doc for marker options..

# format= 'marker|line|color'

```python
plt.plot(xpoints2, ypoints2, 'o:r') 
plt.show()
```

```python
plt.plot(xpoints2, ypoints2, 'o:r', ms = 20)# ms = marker size
plt.show()
```

```python
plt.plot(xpoints2, ypoints2, marker ='o', ms = 20, mec = 'r')# mec = marker edge color
plt.show()
```

```python
plt.plot(xpoints2, ypoints2, marker ='o', ms = 20, mfc = 'r')# mfc = marker face color
plt.show()
```

```python
plt.plot(xpoints2, ypoints2, marker ='o', ms = 20, mec = 'r', mfc = 'r')# edge and face both same color
plt.show()
```

```python
plt.plot(xpoints2, ypoints2, linestyle = 'dashdot')
plt.show()
```

```python
plt.plot(xpoints2, ypoints2, ls = '-.') # alternative
plt.show()
```

```python
plt.plot(xpoints2, ypoints2, color = "g") # line color
plt.show()
```

# multiple plot

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.

plt.plot(x,y)

plt.show()
```

```python
plt.plot(x,y)
plt.title("SALES")
```

```python
plt.scatter(x, y)
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
plt.scatter(x, y, s = [20,50,100]) # size for 1st, 2nd and 3rd dot

plt.show()
```

# bar

```python
x = np.array(["Apple", "Banana", "Cat", "Dog"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color = "red")
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
plt.bar(x,y, width = 0.1)
plt.show()
```

```python
plt.barh(x, y, height = 0.1)
plt.show()
```
